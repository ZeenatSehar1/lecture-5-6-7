1. Build a calculator() function that returns nested functions for add, subtract, multiply, and
divide.
Each returned function should take two parameters and return the result.
Validate division by zero

```python
def calculator():
    def add(a, b):
        return a + b

    def subtract(a, b):
        return a - b

    def multiply(a, b):
        return a * b

    def divide(a, b):
        if b == 0:
            raise ValueError("Cannot divide by zero")
        return a / b

    return add, subtract, multiply, divide

add, sub, mul, div = calculator()

print(add(5, 3))
print(sub(5, 3))
print(mul(3, 3))
print(div(6, 2))
```

. Generator from Dictionary Keys/Values
Write two generators:
dict_keys(d) yields keys
dict_values(d) yields values

```python
def dict_keys(d):
    for key in d:
        yield key


def dict_values(d):
    for value in d.values():
        yield value

data = {"a": 1, "b": 2, "c": 3}

for k in dict_keys(data):
    print(k)

for v in dict_values(data):
    print(v)
```

1. Multi-user Login Validator with Lockout
Create a function create_user_validator(user_dict) that takes a dictionary of usernames
and passwords and returns a closure for validating logins.
Lock each user after 3 incorrect attempts.
Track login attempts per user.
Return appropriate error/success messages.

```python
def create_user_validator(user_dict):
    attempts = {user: 0 for user in user_dict}
    locked = {user: False for user in user_dict}

    def validate_login(username, password):
        if username not in user_dict:
            return "Error: User does not exist."
        if locked[username]:
            return f"Error: User '{username}' is locked due to 3 failed login attempts."
        if user_dict[username] == password:
            attempts[username] = 0
            return f"Success: User '{username}' logged in successfully!"
        else:
            attempts[username] += 1
            if attempts[username] >= 3:
                locked[username] = True
                return f"Error: User '{username}' is now locked due to 3 failed attempts."
            else:
                return f"Error: Incorrect password. Attempt {attempts[username]} of 3."

    return validate_login

users = {
    "alice": "alice123",
    "bob": "bobpass",
    "charlie": "charlie321"
}

validator = create_user_validator(users)

print(validator("alice", "wrong"))
print(validator("alice", "wrong"))
print(validator("alice", "wrong"))
print(validator("alice", "alice123"))
print(validator("bob", "bobpass"))
print(validator("unknown", "123"))
```

2. @log_activity Decorator
Write a decorator @log_activity that:
Logs the function name, arguments, and return value
Timestamp each call
Apply this decorator to multiple business functions.

```python
from functools import wraps
from datetime import datetime

def log_activity(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        result = func(*args, **kwargs)
        print(f"[{timestamp}] Function '{func.__name__}' called with args={args}, kwargs={kwargs}, returned={result}")
        return result
    return wrapper

# Example business functions
@log_activity
def add(a, b):
    return a + b

@log_activity
def multiply(a, b):
    return a * b

@log_activity
def greet(name):
    return f"Hello, {name}!"

add(5, 7)
multiply(3, 4)
greet("Alice")
```
