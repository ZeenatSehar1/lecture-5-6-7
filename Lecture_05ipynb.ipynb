Given a list l = list(range(1, 11)) and a check_even_odd(n) function already
created:
Create two new lists: one for even numbers and one for odd
numbers using the function.
Print both lists.

```python
def check_even_odd(n):
    if n % 2 == 0:
        return "even"
    else:
        return "odd"

l = list(range(1, 11))

even_list = [n for n in l if check_even_odd(n) == "even"]
odd_list  = [n for n in l if check_even_odd(n) == "odd"]

print("Even numbers:", even_list)
print("Odd numbers:", odd_list)
```

Create a list from 100 to 150 using list comprehension where prime numbers stay as-is, and
non-primes become "Not Prime".

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

lst = [i if is_prime(i) else "Not Prime" for i in range(100, 151)]
print(lst)
```

Write a function prime_upto(n) that returns all prime numbers up to n.
Include input validation for type and value.

```python
def prime_upto(n):
    if not isinstance(n, int):
        raise TypeError("n must be an integer")
    if n < 2:
        raise ValueError("n must be >= 2")

    primes = []
    for num in range(2, n + 1):
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
    return primes

prime_upto(10)
```

Write a function that accepts a list of numbers and returns a dictionary of:
total
average
max
min
all even numbers (in a list)

```python
def analyze_numbers(lst):
    if type(lst) != list:
        raise TypeError("Input must be a list")
    if len(lst) == 0:
        raise ValueError("List cannot be empty")

    for i in lst:
        if type(i) not in (int, float):
            raise TypeError("All elements must be numbers")

    total = sum(lst)
    average = total / len(lst)
    maximum = max(lst)
    minimum = min(lst)
    even_numbers = [i for i in lst if type(i) == int and i % 2 == 0]

    return {
        "total": total,
        "average": average,
        "max": maximum,
        "min": minimum,
        "even_numbers": even_numbers
    }

print(analyze_numbers([1, 2, 3, 4, 5, 6]))
```

**Email Filtering System**
You have a list of email strings. Extract all emails from
a specific domain (e.g.,
"@company.com") using list
comprehension. Then count how many unique users
(before the @) there are using set comprehension.
Email Filtering System
emails = ["john@company.com"
,
"alice@gmail.com"
,
"bob@company.com"
,
"john@company.com"]

```python
emails = ["john@company.com",
          "alice@gmail.com",
          "bob@company.com",
          "john@company.com"]

unique_emails = set(emails)

company_emails = []
personal_emails = []

for email in unique_emails:
    if email.endswith("@company.com"):
        company_emails.append(email)
    else:
        personal_emails.append(email)

print("Company Emails:", company_emails)
print("Personal Emails:", personal_emails)
```
